```python
from ipyleaflet import (Map,DrawControl)
myMap = Map(center=[27.571333, 76.608594],zoom=16)
dc = DrawControl(circle={'shapeOptions':{'color':'red'}})
myMap.add_control(dc)
myMap
```

```python
import folium
import folium.plugins as plugins
```

```python
map_measure = folium.Map([27.571333, 76.608594], zoom_start=5)
measure_control = plugins.MeasureControl(position='topleft',
                                        active_color='red',
                                        completed_color='red',
                                        primary_length_unit='kilometers')

map_measure.add_child(measure_control)
map_measure
```

```python
map_dual = plugins.DualMap(location=[16.865662293310375,74.62841033935547],tiles=None,zoom_start=4)

folium.TileLayer('Stamen Terrain').add_to(map_dual)
folium.TileLayer('CartoDB Positron').add_to(map_dual)
folium.LayerControl().add_to(map_dual)

measure_control = plugins.MeasureControl(position='topleft',
                                        active_color='red',
                                        completed_color='red',
                                        primary_length_unit='kilometers')

draw = plugins.Draw(export=True)

map_dual.add_child(measure_control)
draw.add_to(map_dual)
map_dual
```
